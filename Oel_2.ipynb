```python
import sympy as sp
import easyocr
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
from sympy import *

# ==========================================
# Initialize
# ==========================================

x = symbols('x')

reader = easyocr.Reader(['en'])

# ==========================================
# OCR Function
# ==========================================

def read_handwritten_equation():

    print("Upload equation image...")

    uploaded = files.upload()

    filename = list(uploaded.keys())[0]

    result = reader.readtext(filename)

    equation = ""

    for item in result:
        equation += item[1]

    print("\nDetected Equation:")
    print(equation)

    return equation

# ==========================================
# Graph Function
# ==========================================

def plot_graph(expr):

    try:

        expr = expr.replace("^","**")

        func = sp.lambdify(x, sp.sympify(expr))

        X = np.linspace(-10,10,400)

        Y = func(X)

        plt.figure(figsize=(8,5))
        plt.plot(X,Y)

        plt.title("Function Graph")
        plt.xlabel("x")
        plt.ylabel("y")
        plt.grid()

        plt.show()

    except:
        print("Cannot plot graph")

# ==========================================
# Solve Equation
# ==========================================

def solve_equation(equation):

    try:

        equation = equation.replace("^","**")

        left,right = equation.split("=")

        eq = Eq(sympify(left),sympify(right))

        solution = solve(eq,x)

        print("\nSolution:")
        print(solution)

        try:
            print("\nFactorized Form:")
            print(factor(sympify(left)-sympify(right)))
        except:
            pass

    except:
        print("Invalid equation")

# ==========================================
# Differentiate
# ==========================================

def differentiate(expr):

    expr = expr.replace("^","**")

    derivative = diff(sympify(expr),x)

    print("\nDerivative:")
    print(derivative)

# ==========================================
# Integrate
# ==========================================

def integrate_expression(expr):

    expr = expr.replace("^","**")

    integral = integrate(sympify(expr),x)

    print("\nIntegral:")
    print(integral,"+ C")

# ==========================================
# Main Chatbot
# ==========================================

while True:

    print("\n==========================")
    print("AI Mathematics Chatbot")
    print("==========================")

    print("1. Solve Equation")
    print("2. Differentiate")
    print("3. Integrate")
    print("4. Plot Graph")
    print("5. Handwritten Equation Recognition")
    print("6. Exit")

    choice = input("\nSelect Option: ")

    # --------------------------------------

    if choice=="1":

        equation = input(
        "\nEnter equation (example x^2-5*x+6=0): ")

        solve_equation(equation)

    # --------------------------------------

    elif choice=="2":

        expr = input(
        "\nEnter expression: ")

        differentiate(expr)

    # --------------------------------------

    elif choice=="3":

        expr = input(
        "\nEnter expression: ")

        integrate_expression(expr)

    # --------------------------------------

    elif choice=="4":

        expr = input(
        "\nEnter function: ")

        plot_graph(expr)

    # --------------------------------------

    elif choice=="5":

        equation = read_handwritten_equation()

        solve_equation(equation)

    # --------------------------------------

    elif choice=="6":

        print("Goodbye!")
        break

    # --------------------------------------

    else:

        print("Invalid Option")
```
